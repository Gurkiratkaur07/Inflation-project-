```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def inflation_impact_calculator():
    # User inputs
    current_price = float(input("Enter the current price of the item: "))
    inflation_rate = float(input("Enter the expected annual inflation rate (in %): ")) / 100
    years = int(input("Enter the number of years: "))

    # Calculate future prices
    years_array = np.arange(0, years + 1)
    future_prices = current_price * (1 + inflation_rate) ** years_array

    # Create a DataFrame for better visualization
    data = pd.DataFrame({
        'Year': years_array,
        'Future Price': future_prices
    })

    # Print the DataFrame
    print("\nFuture Prices over the years:")
    print(data)

    # Plotting the future prices
    plt.figure(figsize=(10, 5))
    plt.plot(data['Year'], data['Future Price'], marker='o')
    plt.title('Future Price of the Item Over Time')
    plt.xlabel('Years')
    plt.ylabel('Future Price')
    plt.xticks(years_array)
    plt.grid()
    plt.show()

if __name__== "__main__":
    inflation_impact_calculator()
```
